```python
df['publish_time'] = pd.to_datetime(df['publish_time'], errors = 'coerce') # Tarih veri tütüne dönüştürdük
```

- Veri setini yayın yılına göre parçalara bölebilirsiniz

```python
df_2020 = df[df['publish_time'].dt.year == 2020]
df_2021 = df[df['publish_time'].dt.year == 2021]
```

```python
df_2020
```

- Belirli anahtar kelimeleri içeren makaleleri ayırabilirsiniz

```python
df_vaccine = df[df['title'].str.contains("vaccine", case=False, na=False)]
df_treatment = df[df['title'].str.contains("treatment", case=False, na=False)]
```

- Belirli bir dergide yayınlanmış makaleleri ayırabilirsiniz

```python
df_lancet = df[df['journal']=='The Lancet']
df_nature = df[df['journal']=='Nature']
```

```python
df['publish_time'].dt.year.value_counts().sort_index().plot(kind='bar')
plt.title("Yıl bazında yayın sayısı")
plt.xlabel("Yıl")
plt.ylabel("Yayın sayısı")
plt.show()
```

```python
year_range = (df['publish_time'].dt.year >= 2018) & (df['publish_time'].dt.year <= 2023)
df_filtered = df[year_range]
```

```python
df_filtered['publish_time'].dt.year.value_counts().sort_index().plot(kind='bar')
plt.title('2019-2021 Yılları Arasında Yayın Sayısı')
plt.xlabel('Yıl')
plt.ylabel('Yayın Sayısı')
plt.show()
```
